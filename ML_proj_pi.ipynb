````python
#normalisation des donner
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler(feature_range=(0,1))
columns_to_scale = ['Location_Easting_OSGR', 'Location_Northing_OSGR', 'Longitude',
       'Latitude', 'Police_Force', 'Number_of_Vehicles',
       'Number_of_Casualties', 'Day_of_Week', 'Local_Authority_(District)',
       'Local_Authority_(Highway)', '1st_Road_Class', '1st_Road_Number',
       'Road_Type', 'Speed_limit', 'Junction_Detail', 'Junction_Control',
       '2nd_Road_Class', '2nd_Road_Number',
       'Pedestrian_Crossing-Human_Control',
       'Pedestrian_Crossing-Physical_Facilities', 'Light_Conditions',
       'Weather_Conditions', 'Road_Surface_Conditions',
       'Special_Conditions_at_Site', 'Carriageway_Hazards',
       'Urban_or_Rural_Area', 'Did_Police_Officer_Attend_Scene_of_Accident']
scaler.fit(data[columns_to_scale])
scaled_columns = scaler.transform(data[columns_to_scale])
data[columns_to_scale] = scaled_columns
````

````python
#balancing the data
y = data['Accident_Severity']
X = data.drop(columns = ['Accident_Severity'])
na_counts = y.isna().value_counts()

print(na_counts)
````

````python
#distribution initiale

print("Class distribution before SMOTE:")
print(y.value_counts())

# Apply SMOTE to oversample the minority class
from imblearn.over_sampling import SMOTE

smote = SMOTE(sampling_strategy='auto', random_state=42)
X, y = smote.fit_resample(X, y)

# distribution aprées oversampling
print("\nClass distribution after SMOTE:")
print(pd.Series(y).value_counts())
````

````python
oversampled = pd.concat([X, y], axis=1)

sns.countplot(x='Accident_Severity', data=oversampled )
plt.title('Outcome Count Plot')
plt.show()
````

````python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
def metrics_calculator(y_test, ypred, model_name):
    '''
    This function calculates all desired performance metrics for a given model.
    '''
    result = pd.DataFrame(data=[accuracy_score(y_test, ypred),
                                precision_score(y_test, ypred, average='macro'),
                                recall_score(y_test, ypred, average='macro'),
                                f1_score(y_test, ypred, average='macro')],
                          index=['Accuracy','Precision','Recall','F1-score'],
                          columns = [model_name])
    return result
````

````python
r_result = metrics_calculator(y_test, ypred, 'Random')
r_result
````

modele svm

````python
from sklearn.svm import SVC


# Définir le modèle SVM
svm_model = SVC()

# Définir les hyperparamètres à rechercher
parameters = {
    'kernel': ['linear', 'rbf'],  # Types de noyaux à essayer
    'C': [0.1, 1, 10],             # Paramètre de régularisation
}
````

````python
# Créer une instance de GridSearchCV pour rechercher les meilleurs hyperparamètres
grid_search = GridSearchCV(svm_model, parameters, cv=5, scoring='accuracy')

# Ajuster le modèle aux données d'entraînement
grid_search.fit(X_train, y_train)
````

````python
# Obtenir le meilleur modèle après la recherche par grille
best_svm_model = grid_search.best_estimator_

# Utiliser le meilleur modèle pour prédire sur les données de test
y_pred_svm = best_svm_model.predict(X_test)
````

````python
print(svm_model.score(X_train,y_train))
print(svm_model.score(X_test,y_test))
````

````python
# Calculer les mesures de performance
accuracy = accuracy_score(y_test, y_pred_svm)
precision = precision_score(y_test, y_pred_svm, average='macro')
recall = recall_score(y_test, y_pred_svm, average='macro')
f1 = f1_score(y_test, y_pred_svm, average='macro')
````

````python
# Afficher les résultats
svm_result = pd.DataFrame(data=[accuracy, precision, recall, f1],
                          index=['Accuracy', 'Precision', 'Recall', 'F1-score'],
                          columns=['SVM'])

print(svm_result)
````
